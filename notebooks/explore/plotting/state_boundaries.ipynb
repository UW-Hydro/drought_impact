```python
import pandas as pd
import geopandas as gpd
```

```python
path = '/pool0/home/steinjao/data/drought/drought_impact/data/geometry/tl_2022_us_state.shp'
```

```python
df = gpd.read_file(path)
df.to_crs("EPSG:4326");
```

```python
df.plot()
```

```python
non_continental = ['HI','VI','MP','GU','AK','AS','PR']
us49 = df
for n in non_continental:
    us49 = us49[us49.STUSPS != n]
```

```python
us49.boundary.plot()
```
