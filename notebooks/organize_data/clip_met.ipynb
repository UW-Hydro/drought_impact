```python
burn_idx = xr.open_mfdataset(f'{met_path}/burn_idx/*')
burn_idx
```

```python
def clip_105w(data, t_var='day' ):
    data_105w = data.sel(lon=slice(-105))
    data_105w = data_105w.rename({t_var:'time'})
    data_105w_5070 = data_105w.rio.write_crs('EPSG:4326').rio.reproject('EPSG:5070')
    try:
        del data_105w_5070.attrs['grid mapping']
    except:
        pass
    return data_105w_5070.to_dataset()
```

```python
burn_idx_105w_5070_ds = clip_105w(burn_idx['burning_index_g'])
```

```python
burn_idx_105w_5070_ds.attrs = {
    'compilation':'a. stein, 1/19/2023',
    'description' : 'Burn Index G (BI-G)',
    'clipping':'This selection has been clipped to everything west of longitude 105 degrees within CONUS. EPSG:5070 was picked to preserve area for future computations.'
}
```
