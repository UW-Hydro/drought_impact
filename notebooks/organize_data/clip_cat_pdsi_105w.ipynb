```python
dm_path = '/pool0/home/steinadi/data/drought/drought_impact/data/drought_measures'
pdsi = xr.open_dataset(f'{dm_path}/pdsi/pdsi.nc')
```

```python
pdsi_105w = pdsi['pdsi'].sel(lon=slice(-105))
```

```python
pdsi_105w = pdsi_105w.rename({'day':'time'})
pdsi_105w
```

```python
pdsi_105w_5070 = pdsi_105w.rio.write_crs('EPSG:4326').rio.reproject('EPSG:5070')
```

```python
del pdsi_105w_5070.attrs['grid_mapping']
```

```python
pdsi_105w_5070_ds = pdsi_105w_5070.to_dataset()
```

```python
attrs = {
    'compilation':'a. stein, 1/19/2023',
    'description' : 'Palmer Drought Severity Index (PDSI) gathered from various time windows, (specified by the value following the underscore), categorized into USDM categories ndrought.compare.dm_to_usdmcat function.',
    'clipping':'This selection has been clipped to everything west of longitude 105 degrees within CONUS. EPSG:5070 was picked to preserve area for future computations.'
}
```

```python
pdsi_105w_5070_ds.attrs = attrs
```

```python
pdsi_105w_5070_ds.to_netcdf(f'{dm_path}/pdsi/CONUS_105W/pdsi.nc')
```
